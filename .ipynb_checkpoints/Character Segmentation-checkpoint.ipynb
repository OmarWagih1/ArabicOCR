```python
import cv2
import numpy as np
import os, glob, re
import shutil
numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts
results_path = 'C:/D/PDFS/College/4th year/1st term/Pattern Recognition/Project/Dataset/results'
char_path = 'C:/D/PDFS/College/4th year/1st term/Pattern Recognition/Project/Dataset/chars'
char_folder_path = 'C:/D/PDFS/College/4th year/1st term/Pattern Recognition/Project/Dataset/'

dx = [-1, 1, 0, 0, 1, -1, 1, -1]
dy = [0, 0, -1, 1, 1, -1, -1, 1]
#dx = [-1, 1, 0, 0]
#dy = [0, 0, -1, 1]
```

```python
def is_valid(row, col, img_shape):
    if row < 0 or row >= img_shape[0] or col < 0 or col >= img_shape[1]:
        return False
    return True
```

```python
def dfs(img, row, col, visited, remove):
    visited[row, col] = True
    count = 1
    for i in range(len(dx)):
        new_row = row + dx[i]
        new_col = col + dy[i]
        if is_valid(new_row, new_col, img.shape) and not visited[new_row, new_col] and img[new_row, new_col] == 255:
            new_count = dfs(img, new_row, new_col, visited, remove)
            count += new_count
    if remove:
        img[row, col] = 0
    return count
```

```python
def remove_dots(img, baseline_idx, th, file_name): #th is the minimum number of pixels for it to not be a dot or hamza
    bottom_baseline_range = 1
    top_baseline_range = 3
    top_baseline = baseline_idx - top_baseline_range
    bottom_baseline = baseline_idx + bottom_baseline_range
    desired_lower_region_size = 10
    desired_higher_region_size = 6
    top_region_size = top_baseline if top_baseline - desired_higher_region_size < 0 else top_baseline - desired_higher_region_size
    bot_region_size = img.shape[0] - bottom_baseline + 1 if bottom_baseline + desired_lower_region_size + 1 > img.shape[0] else desired_lower_region_size
    upper_img = img[top_baseline - top_region_size:top_baseline, 0: img.shape[1]]
    lower_img = img[bottom_baseline + 1: (bottom_baseline + bot_region_size + 1), 0: img.shape[1]]
    cv2.imwrite(file_name + 'upper.png', upper_img)
    cv2.imwrite(file_name + 'lower.png', lower_img)
    #cv2.imshow("region_img", upper_img)
    img_regions = [upper_img, lower_img]
    img_baselines = [top_baseline, bottom_baseline]
    img_region_sizes = [top_region_size, bot_region_size]
    for i, img_region in enumerate(img_regions):
        print('iteration: ' + str(i))
        visited = np.zeros(img_region.shape)
        baseline = img_baselines[i]
        region_size = img_region_sizes[i]
        for row in range(img_region.shape[0]):
            for col in range(img_region.shape[1]):
                if (img_region[row, col] != 255 or visited[row, col]):
                    continue
                visited_copy = np.copy(visited)
                count = dfs(img_region, row, col, visited, False)
                print(count)
                if count <= th:
                    _ = dfs(img_region, row, col, visited_copy, True)
                    if (i == 0):
                        img[baseline - region_size: baseline, 0: img.shape[1]] = img_region
                    else:
                        img[baseline + 1: baseline + region_size, 0: img.shape[1]] = img_region
    
```

```python
def segment_characters(img_name, folder_count, img_count):
    file_name = char_path + '/' + str(folder) + '/result' + str(img_count) + '/' + 'result' + str(folder_count) + '-' + str(img_count)
    img = cv2.imread(img_name, 0)  #0 parameter means read it in grayscale
    _, img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV|cv2.THRESH_OTSU)
    print(img.shape)
    img_copy = img.copy()
    vertical_profile = cv2.reduce(img, 1, cv2.REDUCE_SUM, dtype=cv2.CV_32S) #2nd param = 1 means column sum
    #print(vertical_profile)
    baseline_idx = np.argmax(vertical_profile)
    remove_dots(img_copy, baseline_idx, 8, file_name)
    H, W = img.shape
    #cv2.line(img_copy, (0,top_baseline), (W, top_baseline), (0,0,255), 1)
    #cv2.line(img_copy, (0,bottom_baseline), (W, bottom_baseline), (0,0,255), 1)
    #cv2.imwrite(file_name +".png", img_copy)
    cv2.imwrite(file_name + "_no_dots.png", img_copy)
    cv2.imwrite(file_name + ".png", img)
    #print(baseline_idx)
```

```python
if os.path.exists(char_folder_path + 'chars'):
    shutil.rmtree(char_folder_path + 'chars', ignore_errors=True)
os.mkdir(char_folder_path + 'chars')
os.chdir(results_path)
for folder_count, folder in enumerate(os.listdir()):
    if (folder_count >= 1):
        break
    os.mkdir(char_path + '/' + str(folder))
    print(folder)
    os.chdir(folder)
    for img_count, img_name in enumerate(sorted(os.listdir(), key=numericalSort)):
        if (img_count >= 20):
            break
        os.mkdir(char_path + '/' + str(folder) + '/result' + str(img_count))
        print(img_name)
        segment_characters(img_name, folder_count, img_count)
    os.chdir('..')
```
